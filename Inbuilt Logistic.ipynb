```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_blobs(centers=4, random_state=42)
```

```python
X.shape, y.shape
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
y
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
model.classes_
```

```python
model.coef_
```

```python
model.intercept_
```

```python
colors = ["Red", "Yellow", "Green", "Orange", "Black"]
```

```python
y_col = [colors[item] for item in y]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_col)
```

```python
import seaborn as sn

sn.scatterplot(X[:, 0], X[:, 1], hue=y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_col)

for index, [[m1, m2], c] in enumerate(zip(model.coef_, model.intercept_)):
    
    x1 = np.linspace(-10, 10, 2)
    x2 = (-m1/m2)*x1 -(c/m2)
    plt.plot(x1, x2, color=colors[index])

plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

```python
model.score(X, y)
```

```python
model.predict_proba(X).argmax(axis=1)
```
