```python
import numpy as np
import statsmodels.api as sm
lowess = sm.nonparametric.lowess
x = np.random.uniform(low=-2*np.pi, high=2*np.pi, size=500)
y = np.sin(x) + np.random.normal(size=len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
data=lowess(y,x)
```

```python
xp=data[:, 0]
```

```python
yp=data[:, 1]
```

```python
plt.scatter(x,y)
plt.scatter(xp,yp)
```
