```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean=[10,20]
covar=[[4,0],[0,6]]
```

```python
data=np.random.multivariate_normal(mean,covar,size=100)#rows =100
```

```python
data.shape
```

```python
plt.scatter(data[:,0],data[:,1],marker="^")
```

```python
data[:,0].mean()
```

```python
data[:,1].mean()
```

```python
mean2=[40,50]
covar2=[[7,0],[0,9]]
```

```python
data2=np.random.multivariate_normal(mean2,covar2,size=100)
```

```python
data2.shape
```

```python
plt.scatter(data2[:,0],data2[:,1],marker="^")
```

```python
plt.scatter(data[:,0],data[:,1],marker="^")
plt.scatter(data2[:,0],data2[:,1],marker="*")
```

```python
d=np.vstack([data,data2])#using vertical stack to combine two data
```

```python
#d
d.shape# when two datas of shape (100,2) combined in vertical stack we get 200 rows and 2 columns
```

```python
plt.scatter(d[:,0],d[:,1],marker="^")
```
