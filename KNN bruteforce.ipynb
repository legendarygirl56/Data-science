```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y= (make_blobs(centers=4,random_state=42))#n_samples=100,features=2 by default
```

```python
X.shape
```

```python
y#it represents the class at which respective x belongs
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
point=np.array([-2,0])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)#here X[:, 0] shows one feature along x axis n the other one other feature along y axis
plt.scatter(point[0], point[1], s=60)#c represent color n s represent size
```

```python
distances = np.sqrt(np.sum((X - point) ** 2, axis=1))
```

```python
distances.argsort()
```

```python
k=20
```

```python
selected = distances.argsort()[:k]
```

```python
y[selected]
```
