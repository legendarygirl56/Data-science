```python
x=np.linspace(-10,10,100)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlim(-100,100)#range
plt.ylim(-100,100)
```

```python
plt.figure()
plt.plot(x,x**2,label="square")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("20 numbers")
plt.xlim(-10,10)
plt.ylim(-100,100)
plt.plot(x,x**3,label="cube")
plt.legend()#bcaz of this we can see square label in small box below

plt.figure()#to construct multiple graph figures
plt.plot(x,x**4,label="fourth")
plt.xlabel("this is x label2")
plt.ylabel("this is y label2")
plt.title("20 numbers again")
plt.xlim(-10,10)
plt.ylim(-100,100)
plt.plot(x,x**5,label="fifth")
plt.legend()
```

```python
x=np.linspace(-10,10,100)
```

```python
plt.figure()
plt.plot(x,np.sin(x), label="sinx")
plt.plot(x,np.cos(x), label="cosx")
plt.plot(x,np.tan(x), label="tanx")
plt.title("graph of sin cos and tan")
plt.legend()
plt.ylim(-4,4)
plt.show()
```

```python
theta= np.linspace(-np.pi,np.pi, 100)
radius = 100
```

```python
#for circle x**2 + y**2 = r
# x=radius*sin(theta)  y=radius*cos(theta)
x=radius*np.sin(theta)
y=radius*np.cos(theta)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,y)
```
